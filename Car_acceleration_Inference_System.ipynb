```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

New Antecedent/Consequent objects hold universe variables and membership functions

```python
distance = ctrl.Antecedent(np.arange(0, 100, 1), 'distance')
delta = ctrl.Antecedent(np.arange(-10, 10, 1), 'delta')
acceleration = ctrl.Consequent(np.arange(-10, 10, 1), 'acceleration')
```

Auto-membership function for as many sets (or categories)

```python
distance.automf(names = ['very small', 'small', 'perfect', 'big', 'very big'])
delta.automf(names = ['shirnking fast', 'shrinking', 'stable', 'growing', 'growing fast'])
```

### Auto membership functions for antecedents

```python
distance.automf(names = ['very small', 'small', 'perfect', 'big', 'very big'])
delta.automf(names = ['shirnking fast', 'shrinking', 'stable', 'growing', 'growing fast'])
```

### Membership functions for Consequent

```python
acceleration['brake hard'] = fuzz.trimf(acceleration.universe, [-10, -10, 0])
acceleration['slow down'] = fuzz.trimf(acceleration.universe, [-10, 0, 0])
acceleration['none'] = fuzz.trimf(acceleration.universe, [-10, 0, 10])
acceleration['speed up'] = fuzz.trimf(acceleration.universe, [0, 10, 10])
acceleration['floor it'] = fuzz.trimf(acceleration.universe, [0, 10, 10])
```
